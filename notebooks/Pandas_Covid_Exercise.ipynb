```
import pandas as pd
import ipywidgets as widgets
from ipywidgets import GridspecLayout, AppLayout, Button, Layout,Output
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import io
import base64
#%unload_ext google.colab.data_table
%load_ext google.colab.data_table
from google.colab import data_table
df = pd.read_csv("https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-andamento-nazionale/dpc-covid19-ita-andamento-nazionale.csv")
df["data"] = [el[0:10] for el in df["data"].values.tolist()]
#display(data_table.DataTable(df, include_index=True, num_rows_per_page=10))
```

```
# deceduti: 35597, totale_casi : 284796, dimessi_guariti: 212432, totale_positivi: 36767
def get_increment(df,select):
    incremento = ( df.iloc[-1,:][select] - df.iloc[-2,:][select] ) .to_dict()
    return incremento
def get_status(df,select):
    data = ( df.iloc[-1,:][select]) .to_dict()
    return data  
def get_fig(df,select,title):
    with plt.style.context("seaborn"):
        #print(plt.style.available)
        fig, ax = plt.subplots(1, 1, figsize=(15, 8))
        df.set_index("data").plot(y=select, kind="line", stacked=False,title=title,ax=ax)
        #filename = name+".png"
        #plt.savefig(filename)
        #plt.close()
        #return filename

        # Encoding
        s = io.BytesIO()
        plt.savefig(s, format='png', bbox_inches="tight")
        plt.close()
        s = base64.b64encode(s.getvalue()).decode("utf-8").replace("\n", "")
        return '<img align="left" src="data:image/png;base64,%s">' % s
def img_widgt(htmlbase64):
    #file = open(name, "rb")
    #image = file.read()
    wi = widgets.HTML(value=htmlbase64,layout=Layout(height='auto',  width='48%'))
    return wi
def plot_fig(df,select,title):
    n = df["data"].values.tolist()
    fig = go.Figure()
    for name in select:
        m = df[name]
        fig.add_trace(go.Scatter(x=n, y=m,
                      mode='lines',#mode='lines+markers',
                      name=name))
    fig.update_layout(
      hovermode = "x",
      paper_bgcolor = "rgb(0,0,0)" ,
      plot_bgcolor = "rgb(10,10,10)" , 
      title=dict(
          x = 0.5,
          text = title,
          font=dict(
              size = 20,
              color = "rgb(255,255,255)"
          )
      )
    )
    return fig

# Incremento/Status Widgets dynami creation
select = ["deceduti","totale_casi","dimessi_guariti","variazione_totale_positivi"]
incremento = get_increment(df,select)
data = get_status(df,select)
#print(data)
#print(incremento)

perc=str(100.0/len(select))+ "%"
llistOggi = []
llistIncremento = []
for name in list(data.keys()):
    wOggi = widgets.Label(value=name + " : " + str(data[name]),layout=Layout(height='auto',
                                                  width=perc,border='3px solid red',
                                                  padding='5px'))
    wIncremento = widgets.Label(value="Incremento --->" + name + " : " + str(incremento[name]),layout=Layout(height='auto',
                                                  width=perc,border='3px solid red',
                                                  padding='5px'))        
    llistOggi.append(wOggi)
    llistIncremento.append(wIncremento)

woggi = widgets.HBox(llistOggi,layout=Layout(height='auto',  width='auto'))
wincremento = widgets.HBox(llistIncremento,layout=Layout(height='auto',  width='auto'))

# Table Creation
table = Output(layout=Layout(height='auto', width='auto'))
with table:
    display(data_table.DataTable(df, include_index=True, num_rows_per_page=10))

# Figure Matplotlib/Pandas Creation
fig_vp = get_fig(df,select=["variazione_totale_positivi"],title="Variazione Totale Positivi")
fig_an = get_fig(df,select=["deceduti","dimessi_guariti","totale_positivi"],title="Andamento Nazionale")
woutfig_vp = img_widgt(fig_vp)
woutfig_an = img_widgt(fig_an)
wfig = widgets.HBox([woutfig_vp,woutfig_an],layout=Layout(height='auto',  width='auto'))

# Figure interactive
wfig_vp_int = Output(layout=Layout(height='auto', width='49%'))
wfig_an_int = Output(layout=Layout(height='auto', width='49%'))
fig_vp_int = plot_fig(df,["variazione_totale_positivi"], title="Variazione Totale Positivi")
fig_an_int = plot_fig(df,["deceduti","dimessi_guariti","totale_positivi"],title="Andamento Nazionale")
with wfig_vp_int:
    display(fig_vp_int)
with wfig_an_int:
    display(fig_an_int)
wifig = widgets.HBox([wfig_vp_int,wfig_an_int],layout=Layout(height='auto',  width='auto'))

# Display
stack = widgets.VBox([woggi,wincremento,table,wfig,wifig])
display(stack)
```
